```python
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3]
y = [2, 4, 1]
plt.plot(x, y)
plt.savefig('q1.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [2, 4, 6, 8, 10]
y = [6, 7, 8, 2, 4]
plt.bar(x, y, label="Bars1")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Bar Graph')
plt.legend()
plt.savefig('q2.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
slices = [7, 2, 2, 13]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['c', 'm', 'r', 'b']
plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow=True,
        explode=(0, 0.1, 0, 0),
        autopct='%1.1f%%'
        )
plt.title('Pie Plot')
plt.savefig('q3.png')
plt.show()
```
